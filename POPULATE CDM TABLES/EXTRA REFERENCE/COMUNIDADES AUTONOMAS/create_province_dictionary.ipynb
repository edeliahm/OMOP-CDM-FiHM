```python
import pandas as pd
```

```python
df = pd.read_excel(r'Tabla Provincias.xls')

del df['Fecha Fin']
del df['Fecha Inicio']

df.head()
```

```python
cols = ['nombre_pais', 'nombre_com_autonoma', 'nombre_provincia']

for col in cols:
    df[col] = df[col].str.title()

df.head()
```

```python
# export 
df.to_csv('diccionario_provincias.csv', index  = False)
```
